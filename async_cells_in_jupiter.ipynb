# Асинхронные ячеки

## Инициализация объектов

```python
print("если вызвать import asyncio вначала, то может быть ошибка")
```

```python
import asyncio
from datetime import datetime

collected = []

# Observable-like async generator
async def numbers(sleep_time: int = 1):
    for i in range(3):
        await asyncio.sleep(sleep_time)
        yield i

# Фоновая задача, которая "подписывается" на генератор
async def run_numbers():
    global collected
    collected = []
    async for n in numbers(2):
        collected.append(n)
        print(f"{n}: {datetime.now()}")
```

```python

# Создаём задачу в фоне
task = asyncio.create_task(run_numbers())
```

## Если вызвать time.sleep, то это блокирует event loop jupiter и обработка task останавливается

```python
task = asyncio.create_task(run_numbers())
```

```python
import time


print(collected)
time.sleep(10)
print(collected)
```

## Если вызвать await asyncio.sleep(10), то это НЕ блокирует event loop jupiter и обработка task продолжается

```python
task = asyncio.create_task(run_numbers())
```

```python
import time


print(collected)
await asyncio.sleep(10)
print(collected)
```
